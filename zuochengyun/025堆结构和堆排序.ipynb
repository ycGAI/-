```python
class HeapSort:
    def heapify(self, nums, i, size):
        l = i * 2 + 1
        while l < size:
            best = l + 1 if l + 1 < size and nums[l + 1] > nums[l] else l
            best = best if nums[best] > nums[i] else i
            if best == i:
                break
            nums[best], nums[i] = nums[i], nums[best]
            i = best
            l = i * 2 + 1

    def heap_insert(self, nums, i):
        while nums[i] > nums[(i - 1) // 2]:
            nums[i], nums[(i - 1) // 2] = nums[(i - 1) // 2], nums[i]
            i = (i - 1) // 2

    def heap_sort(self, nums):
        n = len(nums)
        for i in range(n):
            self.heap_insert(nums, i)
        size = n
        while size > 1:
            size -= 1
            nums[0], nums[size] = nums[size], nums[0]
            self.heapify(nums, 0, size)
        return nums  # 添加返回值

    def heap_sort2(self, nums):
        n = len(nums)
        for i in range(n - 1, -1, -1):
            self.heapify(nums, i, n)
        size = n
        while size > 1:
            size -= 1
            nums[0], nums[size] = nums[size], nums[0]
            self.heapify(nums, 0, size)
        return nums  # 添加返回值

# 测试代码
a = [1, 43, 4, 5, 1, 5, 6, 7, 2, 6, 7, 99, 2]
s = HeapSort()
r = s.heap_sort(a)  # 或者使用 s.heap_sort2(a)
print(r)  # 输出排序后的数组
```
