“**==**”中比较的是**值**，不是内存地址       
**is**比较的是内存地址       
由于python中字符串驻留机制，相同的字符串通常会共享地址，但是不是绝对的

比较器：定制版的比较工具    
对于一些结构需要自定义的排序方式    
list.sort()和sorted()都不可以自定义比较器    
list.sort()仅用于list，**sorted()**可以用于**任何可迭代对象**       


```python
student_tuples = [
    ('john', 'A', 15),
    ('jane', 'B', 12),
    ('dave', 'B', 10),
]
print(sorted(student_tuples, key=lambda student: student[2]))  #年龄升序排序 
print(sorted(student_tuples, key=lambda student: student[2], reverse=True)) #年龄降序排序
```

```python
class Student:
    def __init__(self, name, grade, age):
        self.name = name
        self.grade = grade
        self.age = age
    def __repr__(self):
        return repr((self.name, self.grade, self.age))

student_objects = [
    Student('john', 'A', 15),
    Student('jane', 'B', 12),
    Student('dave', 'B', 10),
]
print(sorted(student_objects, key=lambda student: student.age))# sort by age
```
